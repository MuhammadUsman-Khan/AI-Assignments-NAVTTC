```python
# importing the dependencies
import pandas as pd
import numpy as np
# for stratifiedshuffle split
from sklearn.model_selection import StratifiedShuffleSplit
# for StandardScaler
from sklearn.preprocessing import StandardScaler
# for logisticregression model
from sklearn.linear_model import LogisticRegression
# for accuracy score
from sklearn.metrics import accuracy_score
```

```python
# loading and observing the dataset
df = pd.read_csv('/content/creditcard.csv')
df.head()
```

```python
# Analyzing the dataset
df.info()
```

```python
# number of rows and columns of the dataset
df.shape
```

```python
# since all values are non-null so we have to proceed to analyze the dataset more
df.describe()
```

```python
# Counting the class columns values
df['Class'].value_counts()
```

```python
# loading the data into X & Y variables
X = df.drop(columns='Class', axis=1)
y = df['Class']
```

0 --> Non-Fraud

1 --> Fraud

```python
# We have only 492 values of fraud so we have to use stratifiedshufflesplit for equal ratio distribution in test and train data
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(df, df['Class']):
  X_strat_train = X.loc[train_index]
  X_strat_test = X.loc[test_index]
  y_strat_train = y.loc[train_index]
  y_strat_test = y.loc[test_index]
```

```python
# checking the ratio of out all fraud values as if they are equally distributed or not
print("Fraud ratio in full data", y.mean().round(4))
print("Fraud ratio in train data", y_strat_train.mean().round(4))
print("Fraud ratio in test data", y_strat_test.mean().round(4))
```

```python
# Hence the ratio is same this means our fraud values are splitted in equal ratio
# Now we have to apply Standard Scaler to our data so our values lies in a standard range and outliers are eliminated automatically
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_strat_train)
X_test_scaled = scaler.transform(X_strat_test)
```

```python
# now storing these numpy arrays into the pandas dataframe
X_df_train_scaled = pd.DataFrame(X_train_scaled, columns=X_strat_train.columns, index=X_strat_train.index)
X_df_test_scaled = pd.DataFrame(X_test_scaled, columns=X_strat_test.columns, index=X_strat_test.index)
```

```python
# now checking these dataframes
X_df_train_scaled.head()
```

```python
X_df_test_scaled.head()
```

**This is extra apart from practice**

```python
# storing the logistic regression model into a variable
model = LogisticRegression()
model.fit(X_df_train_scaled, y_strat_train)
```

```python
# now checking the accuracy based on our training data
X_train_prediction = model.predict(X_df_train_scaled)
X_train_accuracy = accuracy_score(X_train_prediction, y_strat_train)
print("Accuracy of training data is ", X_train_accuracy)
```

```python
# now checking the accuracy based on our testing data
X_test_prediction = model.predict(X_df_test_scaled)
X_test_accuracy = accuracy_score(X_test_prediction, y_strat_test)
print("Accuracy of testing data is ", X_test_accuracy)
```
