```python
# importing the dependencies
import pandas as pd
```

```python
# Loading the datasets
pizza_df = pd.read_csv('/content/pizzas.csv')
pizza_types_df = pd.read_csv('/content/pizza_types.csv', encoding='latin-1')
orders_df = pd.read_csv('/content/orders.csv')
order_details_df = pd.read_csv('/content/order_details.csv')
```

```python
print("Orders Dataframe")
orders_df.head()
```

```python
print("Orders Details Dataframe")
order_details_df.head()
```

```python
# Exploring the Dataframes
print("Pizza Dataframe Info")
print(pizza_df.info())
print("\nShape of Dataframe", pizza_df.shape)
print("\nMissing values in Dataframe")
print(pizza_df.isnull().sum())


print("Pizza Types Dataframe Info")
print(pizza_types_df.info())
print("\nShape of Dataframe", pizza_types_df.shape)
print("\nMissing values in Dataframe")
print(pizza_types_df.isnull().sum())

print("Orders Dataframe Info")
print(orders_df.info())
print("\nShape of Dataframe", orders_df.shape)
print("\nMissing values in Dataframe")
print(orders_df.isnull().sum())

print("Orders Details Dataframe Info")
print(order_details_df.info())
print("\nShape of Dataframe", order_details_df.shape)
print("\nMissing values in Dataframe")
print(order_details_df.isnull().sum())
```

```python
# Using merge method
merged_orders_pizzas = pd.merge(order_details_df, pizza_df, on='pizza_id')
merged_orders_pizzas.head()
```

```python
# Using Groupby method
merged_orders_pizzas.groupby('size')['quantity'].sum()
```

```python
# Using pivot method
pizza_df_pivot = pd.pivot(pizza_df,index='pizza_id', columns='size', values='price')
pizza_df_pivot.head()
```

```python
# Using melt method
pizza_df_pivot_reset = pizza_df_pivot.reset_index()
pizza_df_melted = pd.melt(pizza_df_pivot_reset, id_vars='pizza_id', value_vars=['L', 'M', 'S', 'XL', 'XXL'], var_name='size', value_name='price')
pizza_df_melted.head()
```
